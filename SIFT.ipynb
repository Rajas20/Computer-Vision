```python
import cv2
import matplotlib.pyplot as plt

# Load the two images in grayscale
img1 = cv2.imread('ASW.jpeg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('ASWE.jpeg', cv2.IMREAD_GRAYSCALE)

# Initialize SIFT detector
sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

matches = bf.match(des1, des2)
matches = sorted(matches, key=lambda x: x.distance)

N = 50
matched_img = cv2.drawMatches(img1, kp1, img2, kp2, matches[:N], None, flags=2 )

# Display the image
plt.figure(figsize=(12, 6))
plt.imshow(matched_img)
plt.title('SIFT Keypoint Matches')
plt.axis('off')
plt.show()
```
